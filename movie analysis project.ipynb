```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figurre

%matplotlib  inline
matplotlib.rcparams['figure.figsize']=(12,8) #Aadjust the configuration of the plots we will create 
```

```python
#Read in The Dataset
df=pd.read_csv(r'C:\Users\Oshodi Raymond\Desktop\movies.csv')
df.head()
```

```python
df.dtypes
```

```python
nan_df= df[df.isna().any(axis=1)]
df=df.dropna(how='any')
df.head()
```

```python
##Change the data type of columns 
df['budget']=df['budget'].astype('int64')
df['gross']=df['budget'].astype('int64')
df
```

```python
df['yearcorrect']=df['released'].astype(str).apply(lambda x: x.split (',')[-1])
df.head(1000)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"green"}, line_kws={"color":"red"})
```

```python
df.corr(method='pearson') ## other types of correlation ae kendall, Pearson and Spearman
```

```python
correlation_matrix= df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('correlation martix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df.head()
```

```python
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix= df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('correlation martix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```
